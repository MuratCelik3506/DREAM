```python
import networkx as nx
import numpy as np
import pandas as pd
```

```python


def getFeatureVect(G, name="NoName"):
    values = []
    degree_list = np.array(list(G.degree()), dtype=int)[:,1]
    values.append(name)
    values.append(nx.number_of_nodes(G))
    values.append(nx.number_of_edges(G))
    values.append(nx.density(G))
    values.append(np.max(degree_list))
    values.append(np.mean(degree_list))
    values.append(max(list(nx.core_number(G).values())))
    # print("Halfway Done!")
    values.append(nx.average_clustering(G))
    values.append(nx.diameter(G))
    # print("Almost finished!")
    values.append(nx.average_shortest_path_length(G))
    # print("Making final Calculations!")
    values.append(sum(list(nx.triangles(G).values()))/3)
    # print("Last one.")
    values.append(np.mean(np.array(list(nx.eigenvector_centrality_numpy(G).values()))))
    return np.array(values)


feature_categories = np.array(["Name", "Vertex", "Edges", "Density", "Max. Degree", "Avg. Degree", "Max. k-core", "Avg. Clustering Coeff.", "Diameter", "Average Path Length"
                                , "Total Triangles", "Avg. Eigenvector Centrality"])


```

```python
def getName(typeG,nameG,num):
    return "./data/"+typeG+"/"+nameG+num+".txt"
```

```python
df_PA = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(100):
    nameFile = getName('PA',"PreAT_PA_",str(x))
    G = nx.read_edgelist(nameFile)
    data = getFeatureVect(G,"PA")
    df_PA.loc[i] = data
    i+=1
df_PA
```

```python
df_CL = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(100):
    nameFile = getName('CL',"CHUNGLU_CL_",str(x))
    G = nx.read_edgelist(nameFile)
    data = getFeatureVect(G,"CL")
    df_CL.loc[i] = data
    i+=1
df_CL
```

```python
df_CNFG = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(100):
    nameFile = getName('CNFG',"ConfG_CNFG_",str(x))
    G = nx.read_edgelist(nameFile)
    data = getFeatureVect(G,"CNFG")
    df_CNFG.loc[i] = data
    i+=1
df_CNFG
```

```python
df_GNP = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(100):
    nameFile = getName('GNP',"GNP_GNP_",str(x))
    G = nx.read_edgelist(nameFile)
    data = getFeatureVect(G,"GNP")
    df_GNP.loc[i] = data
    i+=1
df_GNP
```

```python
frames = [df_PA,df_CL,df_GNP,df_CNFG]
```

```python
result = pd.concat(frames,ignore_index=True)
result
```

```python
df_feature = result[['Vertex', 'Edges', 'Density', 'Max. Degree', 'Avg. Degree',
       'Max. k-core', 'Avg. Clustering Coeff.', 'Diameter',
       'Average Path Length', 'Total Triangles',
       'Avg. Eigenvector Centrality']]
```

```python
df_label= result["Name"]
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feature,df_label,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(X_train, y_train)
y_pred = svm_clf.predict(X_test)
knn_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", str(round(knn_acc*100, 2)) + "%")
```

```python
from sklearn import linear_model
clf_svm_l1 = linear_model.SGDClassifier(penalty='l1')
clf_svm_l1.fit(X_train,y_train)
y_pred = clf_svm_l1.predict(X_test)
knn_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", str(round(knn_acc*100, 2)) + "%")
```

```python
clf_svm_l2 = linear_model.SGDClassifier(penalty='l2')
clf_svm_l2.fit(X_train,y_train)
y_pred = clf_svm_l2.predict(X_test)
knn_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", str(round(knn_acc*100, 2)) + "%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
y_pred = knn_clf.predict(X_test)

# Computes and prints the classification accuracy
knn_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", str(round(knn_acc*100, 2)) + "%")
```

```python
from sklearn.linear_model import LogisticRegression

# Uses the SVM classifier to perform classification
reg_clf = LogisticRegression(solver='lbfgs', max_iter=1000)
reg_clf.fit(X_train, y_train)
y_pred = reg_clf.predict(X_test)

# Computes and prints the classification accuracy
score = reg_clf.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Uses the SVM classifier to perform classification
adb_clf = AdaBoostClassifier()
adb_clf.fit(X_train, y_train)
y_pred = adb_clf.predict(X_test)

# Computes and prints the classification accuracy
adb_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", str(round(adb_acc*100, 2)) + "%")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Uses the SVM classifier to perform classification
rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

score = rf_clf.score(X_test, y_test)
print(score)
```
