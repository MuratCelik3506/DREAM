```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
G_Enz_540 = nx.read_edgelist("ENZYMES_g540.edges")
G_Enz_g300 = nx.read_edgelist("ENZYMES_g300.edges")
bn_mouse_brain_1 = nx.read_edgelist("bn-mouse_brain_1.edges")
bn_mouse_brain_1.remove_edges_from(nx.selfloop_edges(bn_mouse_brain_1))
bn_macaque_rhesus_brain_2 = nx.read_edgelist("bn-macaque-rhesus_brain_2.edges")
soc_dolphins = nx.read_edgelist("soc-dolphins.mtx")
soc_firm_hi_tech = nx.read_edgelist("soc-firm-hi-tech.txt",nodetype=int,
  data=(('weight',float),))
bio_celegans_dir = nx.read_edgelist("bio-celegans-dir.edges")
bio_celegans_dir.remove_edges_from(nx.selfloop_edges(bio_celegans_dir))
list_Graph = [G_Enz_540,G_Enz_g300,bn_mouse_brain_1,bn_macaque_rhesus_brain_2,soc_dolphins,soc_firm_hi_tech,bio_celegans_dir]
list_Graph_name = ["G_Enz_540","G_Enz_g300","bn_mouse_brain_1","bn_macaque_rhesus_brain_2","soc_dolphins","soc_firm_hi_tech","bio_celegans_dir"]
```

```python
print(nx.is_connected(G_Enz_540))
print(nx.is_connected(G_Enz_g300))
print(nx.is_connected(bn_mouse_brain_1))
print(nx.is_connected(bn_macaque_rhesus_brain_2))
print(nx.is_connected(soc_dolphins))
print(nx.is_connected(soc_firm_hi_tech))
print(nx.is_connected(bio_celegans_dir))
```

```python


def getFeatureVect(G, name="NoName"):
    values = []
    degree_list = np.array(list(G.degree()), dtype=int)[:,1]
    values.append(name)
    values.append(nx.number_of_nodes(G))
    values.append(nx.number_of_edges(G))
    values.append(nx.density(G))
    values.append(np.max(degree_list))
    values.append(np.mean(degree_list))
    values.append(max(list(nx.core_number(G).values())))
    # print("Halfway Done!")
    values.append(nx.average_clustering(G))
    values.append(nx.diameter(G))
    # print("Almost finished!")
    values.append(nx.average_shortest_path_length(G))
    # print("Making final Calculations!")
    values.append(sum(list(nx.triangles(G).values()))/3)
    # print("Last one.")
    values.append(np.mean(np.array(list(nx.eigenvector_centrality_numpy(G).values()))))
    return np.array(values)


feature_categories = np.array(["Name", "Vertex", "Edges", "Density", "Max. Degree", "Avg. Degree", "Max. k-core", "Avg. Clustering Coeff.", "Diameter", "Average Path Length"
                                , "Total Triangles", "Avg. Eigenvector Centrality"])


```

```python

df = pd.DataFrame(columns=feature_categories)
i = 0

for G in list_Graph:
    print(list_Graph_name[i])
    data = getFeatureVect(G,list_Graph_name[i])    
    df.loc[i] = data
    i+=1

with open("features.csv", "w", newline='') as f:
    f.write(df.to_csv())
    f.close()
```

```python
train_Data = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(20,250,10):
    for y in range(0,10,1):
        print(x," ", y)
        g = nx.erdos_renyi_graph(x,y/10)
        if nx.is_connected(g):
            data = getFeatureVect(g,"ER")
            train_Data.loc[i] = data
            i+=1



with open("features_train_ER.csv", "w", newline='') as f:
    f.write(train_Data.to_csv())
    f.close()

```

```python
train_Data_PA = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(10,250,10):
    for y in range(1,x,5):
        print(x," ",y)
        g = nx.generators.random_graphs.barabasi_albert_graph(x,y)
        if nx.is_connected(g):
            data = getFeatureVect(g,"PA")
            train_Data_PA.loc[i] = data
            i+=1



with open("features_train_PA.csv", "w", newline='') as f:
    f.write(train_Data_PA.to_csv())
    f.close()


```

```python
train_Data_CL = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(10,280,10):
    print(x)
    z_One = [x/10 for i in range(x)]
    z_Two = [z for z in range(x)]
    g = nx.generators.degree_seq.expected_degree_graph(z_One,selfloops=False)
    if nx.is_connected(g):
        data = getFeatureVect(g,"CL")
        train_Data_CL.loc[i] = data
        i+=1
    gTwo = nx.generators.degree_seq.expected_degree_graph(z_Two,selfloops=False)
    if nx.is_connected(gTwo):
        data = getFeatureVect(gTwo,"CL")
        train_Data_CL.loc[i] = data
        i+=1



with open("features_train_CL.csv", "w", newline='') as f:
    f.write(train_Data_CL.to_csv())
    f.close()


```

```python
train_Data_CFG = pd.DataFrame(columns=feature_categories)
i = 0
for x in range(30,400,20):
    for y in range(10,140):
        print(x," ",y)
        sequence = nx.random_powerlaw_tree_sequence(y, tries=5000)   
        print(sequence)
        g = nx.generators.degree_seq.configuration_model(sequence, create_using=nx.Graph())
        if nx.is_connected(g):
            data = getFeatureVect(g,"CFG")
            train_Data_CFG.loc[i] = data
            i+=1



with open("features_train_CFG.csv", "w", newline='') as f:
    f.write(train_Data_CFG.to_csv())
    f.close()


```
